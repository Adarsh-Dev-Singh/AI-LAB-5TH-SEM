```python
class VacuumCleanerAgent:
    def __init__(self, initial_position):
        self.grid = ['Messy', 'Messy']
        self.position = initial_position
        self.cleaned_count = 0

    def print_grid(self):
        print("Current Grid State:", self.grid)

    def is_clean(self):
        return self.cleaned_count == len(self.grid)

    def clean(self):
        if self.grid[self.position] == 'Messy':
            self.grid[self.position] = 'Neat'
            self.cleaned_count += 1
            print(f"üßπ Cleaned position {self.position}!")
        else:
            print(f"üåü Position {self.position} is already neat!")

    def move(self):
        # Move to the next position (toggle between 0 and 1)
        self.position = 1 - self.position  # Toggle between 0 and 1
        print(f"‚û°Ô∏è Moved to position {self.position}.")

    def run(self):
        while not self.is_clean():
            self.clean()
            self.move()
            self.print_grid()

        print("‚úÖ All positions are neat and tidy!")

if __name__ == "__main__":
    initial_position = int(input("Enter initial position (0 or 1): "))

    if initial_position in [0, 1]:
        vacuum = VacuumCleanerAgent(initial_position)
        vacuum.print_grid()
        vacuum.run()
    else:
        print("üö´ Invalid initial position. Please enter 0 or 1.")
print("Adarsh Dev Singh")
```

```python
import random
def is_cell_empty(row, col):
    return board[row][col] == '.'

def is_board_full():
    for i in range(3):
        for j in range(3):
            if board[i][j] == '.':
                return True
    return False

def display_board():
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_for_winner(symbol):
    for i in range(3):
        if all([board[i][j] == symbol for j in range(3)]):
            return True
        if all([board[j][i] == symbol for j in range(3)]):
            return True
    if board[0][0] == symbol and board[1][1] == symbol and board[2][2] == symbol:
        return True
    if board[0][2] == symbol and board[1][1] == symbol and board[2][0] == symbol:
        return True
    return False

board = [['.', '.', '.'],
         ['.', '.', '.'],
         ['.', '.', '.']]

display_board()

while is_board_full():
    player_row = int(input("Enter row (0-2): "))
    player_col = int(input("Enter column (0-2): "))

    while not is_cell_empty(player_row, player_col):
        print("Cell is occupied, try again.")
        player_row = int(input("Enter row (0-2): "))
        player_col = int(input("Enter column (0-2): "))

    board[player_row][player_col] = 'X'
    display_board()

    if check_for_winner('X'):
        print("Congratulations! You win!")
        break

    if not is_board_full():
        print("It's a draw!")
        break

    print("Computer's turn...")
    comp_row, comp_col = random.randint(0, 2), random.randint(0, 2)

    while not is_cell_empty(comp_row, comp_col):
        comp_row, comp_col = random.randint(0, 2), random.randint(0, 2)

    board[comp_row][comp_col] = 'O'
    display_board()

    if check_for_winner('O'):
        print("Computer wins! Better luck next time.")
        break

    if not is_board_full():
        print("It's a draw!")
        break0

print("ADARSH DEV SINGH")
```
